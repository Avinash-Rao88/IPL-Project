### Q.1: What is the maximum number of matches played by an individual player in a season? 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('IPL_Dataset.csv')
data.head()
```

```python
max_matches = data.groupby('Player')['Mat'].max()
max_player = max_matches.idxmax()
max_matches_count = max_matches.max()
print(f"Maximum number of matches played by an individual player in a season is {max_matches_count},\nPlayer Name: {max_player}")
```

### Q.2: What is the highest average of a player in a season? 

```python
highest_average = data.groupby('Player')['Avg'].max()
max_avg_player = highest_average.idxmax()
max_avg = highest_average.max()
print(f"Highest average of a player in a season is: {max_avg}, \nPlayer Name: {max_avg_player}")
```

### Q.3: What is the average number of matches played by the players in a season? 

```python
average_matches_played = data['Mat'].mean()
print(f'The average number of matches played by the players in a season: {average_matches_played}')
```

### Q.4: Please check the correlation between all the features? 

```python
numeric_data = data.select_dtypes(include=['float64', 'int64'])
correlation = numeric_data.corr()
print("Correlation between features:")
print(correlation)
```

### Q.5: Check the list of players who has an average greater than 50 as well strike rate above 120? 

```python
average = data[(data['Avg'] > 50) & (data['SR'] > 120)]
index_order =  average[['Player', 'Avg', 'SR']].reset_index(drop=True)
print('Players who have an average greater than 50 and a strike rate below 120:')
print(index_order)
```

### Q.6: Please check the list of players who has an average greater than 40 and balls faced above 100? 

```python
average = data[(data['Avg'] > 40) & (data['BF'] > 100)]
index_order =  average[['Player', 'Avg', 'BF']].reset_index(drop=True)
print('List of players who has an average greater than 40 and balls faced above 100:')
print(index_order)
```

### Q.7: Please check the list of players who scored more than 300 runs with an average greater than 45 & strike rate above 130? 

```python
list_of_player = data[(data['Runs'] > 300) & (data['Avg'] > 45) & (data['SR'] > 130)]
index_order =  list_of_player[['Player', 'Avg', 'BF']].reset_index(drop=True)
print('The list of players who scored more than 300 runs with an average greater than 45 & strike rate above 130:')
print(index_order)
```

### Q.9: Players who scored atleast 4 half centuries in this season?

```python
half_century = data[data['50'] >= 4]
index_order = half_century[['Player', '50']].sort_values(by='50', ascending=False).reset_index(drop=True)
print('The list of players who scored at least four half century in this season:')
print(index_order)
```

### Q.10: Check the list of players who hit more than 45 boundaries and more than 10 sixes in this season? 

```python
boundaries = data[(data['4s'] > 45) & (data['6s'] > 10)]
index_order = boundaries[['Player', '4s', '6s']].sort_values(by=['4s', '6s'], ascending=False).reset_index(drop=True)
print('The list of players who hit more than 45 boundaries and more than 10 sixes in this season:')
print(index_order)
```

### Q.11: Plot a histogram of number of matches played in a season by players? 

```python
plt.hist(data['Mat'], bins=10, edgecolor='black', color='green')
plt.title('Number of matches played in a season by players')
plt.xlabel('Number of Matches')
plt.ylabel('Player')
plt.show()
```

### Q.12: Plot the histogram of balls faced by players?

```python
plt.hist(data['BF'], bins=10, edgecolor='black', color='yellow')
plt.title('Balls faced by players')
plt.xlabel('Number of Matches')
plt.ylabel('Player')
plt.show()
```

### Q.13: Top 20 players with most runs in a season? 

```python
most_runs = data.nlargest(20, 'Runs')
most_runs_reset = most_runs.reset_index(drop=True)
print('Top 20 players with most runs in a season:')
print(most_runs_reset[['Player', 'Runs']])
```

### Q.14: Bar chart for top 20 players with best strike rate in a season? 

```python
top_players = data.nlargest(20, 'SR')
plt.figure(figsize=(10, 8))
plt.bar(top_players['Player'], top_players['SR'], color='brown')
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.title('Top 20 Players with Best Strike Rate in a Season')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Q.15: Top 30 players with highest number of sixes? 

```python
highest_number_sixes = data.nlargest(30, '6s')
highest_number_sixes_reset = highest_number_sixes.reset_index(drop=True)
print('Top 30 players with highest number of sixes:')
print(highest_number_sixes_reset[['Player', '6s']])
```

### Q.16: Scatter plot of runs scored by a player v/s balls faced in a season.?


```python
plt.scatter(data['Runs'], data['BF'], alpha=0.5, edgecolor='black', color='orange')
plt.title('Runs scored by a player vs balls faced in a season')
plt.xlabel('Runs Scored')
plt.ylabel('Balls Faced')
plt.show()
```

### Q.17: KDE plot for averages on a season?

```python
sns.kdeplot(data['Avg'], fill=True, edgecolor='black', color='brown')
plt.title('KDE plot for averages on a season')
plt.xlabel('Average')
plt.ylabel('Density')
plt.show()
```
